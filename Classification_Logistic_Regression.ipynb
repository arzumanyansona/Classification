```python
import numpy as np
import pandas as pd
```

```python
data= pd.read_csv("/content/classification.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data_ob=pd.DataFrame(data['ed'])
```

```python
data_ob
```

```python
data1= data.select_dtypes(exclude=['object'])
```

```python
data1
```

```python
target=data["default"]
```

```python
data=data.drop(['default'],axis=1)
```

```python
data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False)
ob_data =ohe.fit_transform(data[['ed']])
ob_data
```

```python
ohe_data=pd.get_dummies(data_ob)
ohe_data.keys()
```

```python
ohe_data=pd.DataFrame(ob_data, columns = ohe_data.keys())
```

```python
ohe_data.head()
```

```python
new_data = pd.concat([data1,ohe_data], axis=1)
```

```python
new_data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaller = StandardScaler()
standard_data=scaller.fit_transform(new_data)
standard_data = pd.DataFrame(standard_data, columns=[new_data.keys()])
```

```python
standard_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(standard_data, target, test_size=0.2)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_pred, y_test)
```

```python
print('Accuracy: %.2f'
      % accuracy_score(y_test,y_pred))

print('Confusion_Matrix:', conf_matrix)


print('Recall: %.2f'
      % recall_score(y_test,y_pred))

print('Precision: %.2f'
      % precision_score(y_test,y_pred))
```
